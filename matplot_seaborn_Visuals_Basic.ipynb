# **Matplot lib and seaborn visualization using Diamonds dataset**

```
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
```

```
#loads data from sns - data set present in sns library
data = sns.load_dataset('diamonds')
```

```
data.columns
```

```
data.shape
```

```
data.head(5)
```

```
data1 =data.sample(30)
```

```
pd.crosstab(data1.cut,data1.clarity).plot(kind ='bar')
```

```
data1.clarity.value_counts().plot(kind='pie')
```

```
plt.scatter(data.carat,data.price)
```

```
plt.hist(data.depth,bins=20,facecolor ="pink")
```

```
plt.violinplot(data.depth)
```

```
plt.boxplot(data.depth,vert = False)
```

***Seaborn***




```
sns.stripplot(data1.cut)
```

```
sns.stripplot(x= data.cut, y= data.price)
```

```
sns.swarmplot(x=data1.clarity,y=data1.price ,hue=data1.cut)
```

```
sns.boxplot(x= data.cut, y= data.price)
```

```
sns.violinplot(x= data.cut, y= data.price)
```

```
sns.violinplot(x= data.cut, y= data.price)
sns.stripplot(x= data.cut, y= data.price)
```

```
sns.jointplot(x= data1.carat, y=data1.price)
```

```
sns.jointplot(x= data1.carat, y=data1.price,kind ='kde')
```

```
sns.pairplot(data1)
```
